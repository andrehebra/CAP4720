```python
import pygame
import os
import time
```

```python
pygame.init()
clock = pygame.time.Clock()
display = pygame.display.set_mode((800,600))
pygame.display.set_caption(title = "Assignment 1: Andre Hebra")

running = True
red = 100
redIncrease = 1

green = 200
greenIncrease = 5

blue = 50
blueIncrease = 10


while running:
    clock.tick(30)
    for event in pygame.event.get():
        if event.type == pygame.QUIT or (event.type == pygame.KEYDOWN and event.key == 27):
            running = False
    display.fill((red,green,blue))

    if (1 > red+redIncrease or red+redIncrease > 254): redIncrease = 0-redIncrease 
    if (1 > green+greenIncrease or green+greenIncrease > 254): greenIncrease = 0-greenIncrease
    if (1 > blue+blueIncrease or blue+blueIncrease > 254): blueIncrease = 0-blueIncrease

    red = red + redIncrease
    green = green + greenIncrease
    blue = blue + blueIncrease
    
    

    pygame.display.flip()

    print("next color values:")
    print(red)
    print(green)
    print(blue)

pygame.quit()
```
